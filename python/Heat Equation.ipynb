```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np
```

```python
fig, ax = plt.subplots(2,2,subplot_kw={"projection": "3d"})

i_x = [0,1]
i_y = [0,1]

mu_vec = [1, 2, 32, 128]


# Make the grid
Z = np.arange(0, -15, -0.1)
t = np.arange(0, 2, 0.1)
Z, t = np.meshgrid(Z, t)

k = 0
for i in i_x:
    for j in i_y:
        # calculate the temperature
        mu = mu_vec[k]
        GenExp = np.array(2*np.pi*1j*t + np.sqrt(np.pi/mu)*(1+1j)*Z)
        T = np.imag(np.exp((GenExp)))
        surf = ax[i_x[i], i_y[j]].plot_surface(t, Z, T, cmap=cm.coolwarm, linewidth=0, antialiased=False)
        ax[i_x[i], i_y[j]].set_xlabel("t")
        ax[i_x[i], i_y[j]].set_ylabel("z")
        ax[i_x[i], i_y[j]].view_init(30)
        ax[i_x[i], i_y[j]].title.set_text(("mu ={} ").format(str(mu)))
        k = k + 1


# GenExp = np.array(2*np.pi*1j*t + np.sqrt(np.pi/mu)*(1+1j)*Z)
# T = np.imag(np.exp((GenExp)))
# #2*np.pi*t - np.sqrt(np.pi/mu)*Z

# # Plot the surface.


# # Customize the z axis.
# # ax[0, 0].set_zlim(-1.01, 1.01)
# # ax[0, 0].zaxis.set_major_locator(LinearLocator(10))

# for angle in range(0, 360):
#     ax.view_init(30, angle)
#     plt.draw()
#     plt.pause(.001)





# mu = 10

# # Make data.
# GenExp = np.array(2*np.pi*1j*t + np.sqrt(np.pi/mu)*(1+1j)*Z)
# T = np.imag(np.exp((GenExp)))
# #2*np.pi*t - np.sqrt(np.pi/mu)*Z

# # Plot the surface.
# surf = ax[1, 0].plot_surface(Z, t, T, cmap=cm.coolwarm,
#                        linewidth=0, antialiased=False)

# mu = 50

# # Make data.
# GenExp = np.array(2*np.pi*1j*t + np.sqrt(np.pi/mu)*(1+1j)*Z)
# T = np.imag(np.exp((GenExp)))
# #2*np.pi*t - np.sqrt(np.pi/mu)*Z

# # Plot the surface.
# surf = ax[1, 1].plot_surface(Z, t, T, cmap=cm.coolwarm,
#                        linewidth=0, antialiased=False)

# mu = 500

# # Make data.
# GenExp = np.array(2*np.pi*1j*t + np.sqrt(np.pi/mu)*(1+1j)*Z)
# T = np.imag(np.exp((GenExp)))
# #2*np.pi*t - np.sqrt(np.pi/mu)*Z

# # Plot the surface.
# surf = ax[0, 1].plot_surface(Z, t, T, cmap=cm.coolwarm,
#                        linewidth=0, antialiased=False)
# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)


plt.show()
```
